 Using the given x and y data, generate a 3D surface plot to visualize the function

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

df = pd.DataFrame(data)
fig = go.Figure(data = [go.Surface(x=x,y=y,z=z)])
fig.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure,output_file,show
x = np.random.randint(1,10,5)
y=np.sin(x)
p = figure(title="Sine Function")
p.circle(x,y)
show(p)
```
